```python
south = data[data["Region"] == "South"]
midwest = data[data["Region"] == "Midwest"]
west = data[data["Region"] == "West"]
Northeast = data[data["Region"] == "Northeast"]
```

```python
import matplotlib.pyplot as plt
ax = plt.axes()
ax.set_facecolor("#41524f")
color = ["#b50cf7" , "#026b2e" , "#0a7abf" , "#bf580a"]
plt.bar(["South" , "Midwest" , "West" , "Northeast"] , [len(south) , len(midwest) , len(west) , len(Northeast)] , color=color)
plt.xlabel("REGION" , color="green")
plt.title("REGION COUNT" , color="blue")
plt.show()
```

```python
data["Region"].value_counts()
```

```python
def change(c):
    if c == "South":
        return 0
    elif c == "Midwest":
        return 1
    elif c == "West":
        return 2
    elif c == "Northeast":
        return 3
```

```python
data["Region"] = data["Region"].apply(change)
data
```
